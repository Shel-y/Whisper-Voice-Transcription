```python
from google.colab import files

uploaded = files.upload()
```

```python
import whisper
import os

# Cargar el modelo base (tiny, base, small, medium o large)
model = whisper.load_model("small")

# Obtener nombre del archivo
audio_file = next(iter(uploaded))

# Transcribir
result = model.transcribe(audio_file, language='Spanish')

texto_transcrito = result["text"]

# Mostrar el texto
print("üéß Transcripci√≥n:")
print(texto_transcrito[:500], "...\n")  # solo mostramos los primeros 500 caracteres
```

```python
from google.colab import files

# Guardar resumen en archivo .txt
with open("texto_transcrito.txt", "w", encoding="utf-8") as f:
    f.write(texto_transcrito)  # o texto_transcrito si quieres toda la transcripci√≥n

# Descargar
files.download("texto_transcrito.txt")
```
